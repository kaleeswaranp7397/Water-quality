```python
# make is_safe -> numeric
df.loc[df['is_safe'] == '#NUM!', 'is_safe']= df['is_safe'].mode()[0]
df['is_safe']= df['is_safe'].astype(int)
```

```python
# clearly its unbalanced, so upsample the minority
majority_df= df[ df['is_safe'] == 0 ]
minority_df= df[ df['is_safe'] == 1 ]

majority_df.shape, minority_df.shape
```

```python
# Clearly its unbalanced, so upsample the minority
from sklearn.utils import resample # Import resample

majority_df= df[ df['is_safe'] == 0 ]
minority_df= df[ df['is_safe'] == 1 ]

majority_df.shape, minority_df.shape

upsampled_minority_df = resample(minority_df, replace=True, n_samples=len(majority_df), random_state=42)

# Reset the index of the upsampled minority df
upsampled_minority_df = upsampled_minority_df.reset_index(drop=True)

# Reset index for majority df too (if necessary)
majority_df = majority_df.reset_index(drop=True)

# Combine the datasets into balanced_df
balanced_df = pd.concat([majority_df, upsampled_minority_df]).reset_index(drop=True)
```

```python
balanced_df= balanced_df.sample(frac=1)
```

```python
# features and labels

X= balanced_df.drop('is_safe', axis= 1)

Y= balanced_df['is_safe']
```

```python
# train test
from sklearn.model_selection import train_test_split # Import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, Y, test_size=0.2)
```

```python
# numeric transformer := imputer + power transformer
from sklearn.pipeline import Pipeline # Import Pipeline
from sklearn.impute import KNNImputer # Import KNNImputer
from sklearn.preprocessing import PowerTransformer # Import PowerTransformer

numeric_transformer= Pipeline(
    [
        ('numeric_imputer', KNNImputer(weights='distance')),
        ('power_transformer', PowerTransformer())
    ]
)
```

```python
# preprocessor
from sklearn.compose import ColumnTransformer # Import ColumnTransformer

preprocessor= ColumnTransformer([
    ('numeric_transformer', numeric_transformer, slice(0, 20))
])
```

```python
# model
from sklearn.linear_model import LogisticRegression # Import LogisticRegression

model= LogisticRegression()
```

```python
# pipeline
from sklearn.decomposition import PCA # Import PCA
from sklearn.pipeline import Pipeline # Import Pipeline

pipe= Pipeline([
    ('preprocessor', preprocessor),
    ('PCA', PCA(n_components=17)),
    ('model', model),
])
```

```python
# fit it
pipe.fit(X_train, y_train)
```

```python
# predict
y_preds= pipe.predict(X_test)
```

```python
# evaluate
```

```python
# evaluate
from sklearn.metrics import accuracy_score # Import accuracy_score

accuracy_score(y_test, y_preds)
```

```python
# evaluate
from sklearn.metrics import accuracy_score # Import accuracy_score
from sklearn.metrics import precision_score # Import precision_score

accuracy_score(y_test, y_preds)

precision_score(y_test, y_preds)
```

```python
# evaluate
from sklearn.metrics import accuracy_score # Import accuracy_score
from sklearn.metrics import precision_score # Import precision_score
from sklearn.metrics import recall_score # Import recall_score
from sklearn.metrics import f1_score # Import f1_score

accuracy_score(y_test, y_preds)

precision_score(y_test, y_preds)

recall_score(y_test, y_preds)

f1_score(y_test, y_preds)
```

```python
# find the best params for PCA
max_acc= 0
best_comps= 0

for i in range(1, 20):

    pipe= Pipeline([
        ('preprocessor', preprocessor),
        ('PCA', PCA(n_components=i)),
        ('model', model),
    ])

    pipe.fit(X_train, y_train)

    y_preds= pipe.predict(X_test)

    acc= accuracy_score(y_test, y_preds)

    if acc > max_acc:
        max_acc= acc
        best_comps= i

    print(f"{i} comps := {acc}")

print(f"\n\nBest comps: {best_comps}, Accuracy: {max_acc}")
```
